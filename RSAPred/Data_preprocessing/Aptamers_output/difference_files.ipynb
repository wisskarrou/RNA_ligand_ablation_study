```python
import pandas as pd

# Charge les datasets
v1 = pd.read_csv(r"C:\Users\cycyc\ModifiedRSAPred\Data_preprocessing\Aptamers_output\Final_dataset_Aptamers_v1.csv")
v2 = pd.read_csv(r"C:\Users\cycyc\ModifiedRSAPred\Data_preprocessing\Aptamers_output\Final_dataset_Aptamers_v2.csv")


# Filtrer les colonnes d'intérêt
v1_subset = v1[["Target_RNA_ID", "name", "pKd"]]
v2_subset = v2[["Target_RNA_ID", "name", "pKd"]]

# Fusionner sur le nom de la molécule
merged = pd.merge(v1_subset, v2_subset, on="name", suffixes=('_v1', '_v2'))

# Afficher quelques exemples où la molécule est la même mais le RNA différent
different_rnas = merged[merged["Target_RNA_ID_v1"] != merged["Target_RNA_ID_v2"]]

# Afficher les résultats
print(different_rnas[["name", "Target_RNA_ID_v1", "Target_RNA_ID_v2", "pKd_v1", "pKd_v2"]])
```
